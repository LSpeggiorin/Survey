### Semantic Scholar

```python
from semanticscholar import SemanticScholar

def loc_articles_SemanticScholar(query, verbose = True):
    q = query

    sch = SemanticScholar()

    results = sch.search_paper(query=query, bulk=True, sort='publicationDate:desc')

    all_results = []
    i = 1
    for item in results:
        all_results.append(item)
        i += 1

    if verbose:
        print(f"Query used: ",query)
        print(f"Number of articles: ",len(all_results))

    SemanticScholar_results = pd.DataFrame([paper.raw_data for paper in all_results])

    return SemanticScholar_results
```

```python
from xploreapi.xploreapi import XPLORE
import pandas as pd
import time
from tqdm.notebook import tqdm
import math

def loc_articles_IEEEXplore(query, api_key, max_results = 4000, verbose = True):
    q = query
    key = api_key

    xquery = XPLORE(api_key)
    xquery.searchField('boolean_text', query)
    xquery.dataType("json")
    xquery.dataFormat("object")

    all_articles = []
    start_record = 1
    articles_per_call = 200 # Maximum allowed articles per API call
    api_calls_made = 0

    # Calculate the total number of expected API calls for tqdm
    total_api_calls = math.ceil(max_results / articles_per_call)

    with tqdm(total=total_api_calls, desc="Fetching IEEE Xplore articles") as pbar:
        while start_record <= max_results:
            current_batch_size = min(articles_per_call, max_results - start_record + 1)

            if current_batch_size <= 0:
                break

            xquery.maximumResults(current_batch_size)
            xquery.startingResult(start_record)

            data = xquery.callAPI()
            api_calls_made += 1
            pbar.update(1)

            if data and "articles" in data:
                batch_articles = data["articles"]
                all_articles.extend(batch_articles)
                start_record += len(batch_articles)
            else:
                # If no articles or error, break the loop
                if verbose:
                    print(f"No more articles found or API error at start_record {start_record}.")
                break

            # Respect the rate limit of 10 calls per second (0.1s delay per call)
            time.sleep(0.1)

    raw_ieee_df = pd.DataFrame(all_articles)

    if verbose:
        print(f"Query used: {query}")
        print(f"Number of articles: {len(raw_ieee_df)}")
        print(f"API calls made: {api_calls_made}")

    return raw_ieee_df
```

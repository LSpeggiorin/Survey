### IEEE Xplore


```python
from xploreapi.xploreapi import XPLORE

def loc_articles_IEEEXplore(query, api_key, max_results = 2000, verbose = True):
    q = query
    key = api_key

    xquery = XPLORE(api_key)
    xquery.searchField('boolean_text', query)
    xquery.maximumResults(max_results)
    xquery.resultsSorting('publication_year','asc')
    xquery.dataType("json")
    xquery.dataFormat("object")

    data = xquery.callAPI()
    raw_ieee_df = pd.DataFrame(data.get("articles", []))

    if verbose:
        print(f"Query used: ",query)
        print(f"Number of articles: ",len(raw_ieee_df))

    return raw_ieee_df
```

```python
import springernature_api_client.meta as meta
from springernature_api_client.utils import results_to_dataframe

def loc_articles_springer(query, api_key, max_results = 2000, verbose = True):
    q = query
    key = api_key

    meta_client = meta.MetaAPI(api_key=key)

    response = meta_client.search(q=q, p=50, s=1, fetch_all=True, is_premium=False)

    df = results_to_dataframe(response)

    if verbose:
        print(f"Query used: ",query)
        print(f"Number of articles: ",len(df))

    return df
```
